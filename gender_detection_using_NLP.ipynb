```
import random
```

```
import nltk
nltk.download('names')
from nltk.corpus import names
```

```
def gender_features(word): 
    return {'last_letter':word[-1]} 
```

```
#makes labels as their gender
labeled_names = ([(name, 'male') for name in names.words('male.txt')]+
             [(name, 'female') for name in names.words('female.txt')]) 
```

```
random.shuffle(labeled_names) 
```

```
#processing the name data
featuresets = [(gender_features(n), gender)  
               for (n, gender)in labeled_names] 
```

```
#divide the result in feature
#sets into training set and test test
train_set, test_set = featuresets[500:], featuresets[:500] 
```

```
#use naivebayesclassification for classification
classifier = nltk.NaiveBayesClassifier.train(train_set) 
  
print(classifier.classify(gender_features('amit'))) 
  
```

```
#print first ten rows of training datasets
classifier.show_most_informative_features(10) 
```
